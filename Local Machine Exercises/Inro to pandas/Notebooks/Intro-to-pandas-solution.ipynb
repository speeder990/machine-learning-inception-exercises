# Filtering

### Find all the observations where temperature is greater than 20

```python
df[df[' _tempm']>20]
```

### Find all those observations where the condition is Haze or Clear

```python
df[(df[' _conds'] == "Haze") | (df[' _conds'] == "Clear")]
```

# Grouping

### Find Minimum, Maximum and Average Temperature and Humidity for all weather conditions

```python
df.groupby(' _conds')[' _tempm',' _hum'].agg(['min','max','mean'])
```

### Find Minimum, Maximum and Average Temperature and Humidity for all wind directions in all weather conditions

```python
df.groupby([' _conds',' _wdire'])[' _tempm',' _hum'].agg(['min','max','mean'])
```
